# Codigo

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Cargar los datos
file_path = '/content/2016.csv'
df = pd.read_csv(file_path)

# Variables a analizar
variables = ['Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom']

# Crear los histogramas y boxplots combinados
for variable in variables:
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Histograma
    axs[0].hist(df[variable], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    axs[0].set_title(f'Histograma de {variable}')
    axs[0].set_xlabel(variable)
    axs[0].set_ylabel('Frecuencia')

    # Boxplot
    axs[1].boxplot(df[variable], patch_artist=True, boxprops=dict(facecolor='lightgreen'))
    axs[1].set_title(f'Boxplot de {variable}')
    axs[1].set_ylabel(variable)

    plt.tight_layout()
    plt.show()
```

# Analisis

Análisis de las distribuciones:
Distribución del Puntaje de Felicidad (Happiness Score):
Histograma: La mayoría de los países tienen un puntaje de felicidad entre 4 y 7, con una ligera concentración entre 5 y 6.
Boxplot: No se observan valores atípicos significativos, y la mediana está cerca de 5.5, lo que indica que aproximadamente la mitad de los países tienen puntajes de felicidad por encima de ese valor.
Distribución del PIB per Cápita (Economy - GDP per Capita):
Histograma: La distribución está sesgada hacia la derecha, indicando que la mayoría de los países tienen un PIB per cápita bajo, pero hay algunos países con PIB per cápita muy alto.
Boxplot: Hay algunos valores elevados pero no hay valores atípicos extremos visibles. La mediana está cerca de 1, lo que sugiere que muchos países tienen un PIB per cápita moderado.
Distribución de la Expectativa de Vida Saludable (Health - Life Expectancy):
Histograma: La mayoría de los países tienen una expectativa de vida saludable entre 0.5 y 0.8, con un pico en torno a 0.6.
Boxplot: El rango intercuartil es bastante amplio, y aunque no hay valores atípicos visibles, existen diferencias significativas en la expectativa de vida saludable entre los países.
Distribución de la Libertad para Tomar Decisiones (Freedom):
Histograma: La libertad para tomar decisiones varía entre 0.2 y 0.6, con una distribución sesgada hacia la derecha.
Boxplot: No se observan valores atípicos extremos, pero hay una variabilidad considerable en los valores, con una mediana alrededor de 0.4.
