# Codigo

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Cargar los datos
file_path = '/content/2016.csv'
df = pd.read_csv(file_path)

# Variables a analizar
variables = ['Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom']

# Crear los histogramas y boxplots combinados
for variable in variables:
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Histograma
    axs[0].hist(df[variable], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    axs[0].set_title(f'Histograma de {variable}')
    axs[0].set_xlabel(variable)
    axs[0].set_ylabel('Frecuencia')

    # Boxplot
    axs[1].boxplot(df[variable], patch_artist=True, boxprops=dict(facecolor='lightgreen'))
    axs[1].set_title(f'Boxplot de {variable}')
    axs[1].set_ylabel(variable)

    plt.tight_layout()
    plt.show()
```

# Analisis

Anlisis de las distribuciones:
Distribucin del Puntaje de Felicidad (Happiness Score):
Histograma: La mayora de los pases tienen un puntaje de felicidad entre 4 y 7, con una ligera concentracin entre 5 y 6.
Boxplot: No se observan valores atpicos significativos, y la mediana est cerca de 5.5, lo que indica que aproximadamente la mitad de los pases tienen puntajes de felicidad por encima de ese valor.
Distribucin del PIB per Cpita (Economy - GDP per Capita):
Histograma: La distribucin est sesgada hacia la derecha, indicando que la mayora de los pases tienen un PIB per cpita bajo, pero hay algunos pases con PIB per cpita muy alto.
Boxplot: Hay algunos valores elevados pero no hay valores atpicos extremos visibles. La mediana est cerca de 1, lo que sugiere que muchos pases tienen un PIB per cpita moderado.
Distribucin de la Expectativa de Vida Saludable (Health - Life Expectancy):
Histograma: La mayora de los pases tienen una expectativa de vida saludable entre 0.5 y 0.8, con un pico en torno a 0.6.
Boxplot: El rango intercuartil es bastante amplio, y aunque no hay valores atpicos visibles, existen diferencias significativas en la expectativa de vida saludable entre los pases.
Distribucin de la Libertad para Tomar Decisiones (Freedom):
Histograma: La libertad para tomar decisiones vara entre 0.2 y 0.6, con una distribucin sesgada hacia la derecha.
Boxplot: No se observan valores atpicos extremos, pero hay una variabilidad considerable en los valores, con una mediana alrededor de 0.4.
