# 2. Cargar los Datos

```python

import pandas as pd

# Cargar datos desde un archivo CSV
df = pd.read_csv('/content/CODING DOJO/retail_sales_dataset.csv')

# Ver las primeras y últimas filas del DataFrame
print("Primeras filas del DataFrame:")
df.head(10)
```

# 3. Exploración Inicial de los Datos


```python
# Imprimir las ultimas 5 filas
print("Últimas filas del DataFrame:")
print(df.tail(5))

# Obtener información general sobre el DataFrame
print("Información del DataFrame:")
print(df.info())

# Generar estadísticas descriptivas
print("Estadísticas descriptivas del DataFrame:")
print(df.describe())
```

# 4. Limpieza de Datos

```python
# Eliminar filas que contengan datos faltantes
df_sin_nulos_filas = df.dropna()
print("\n DataFrame sin filas con valores nulos:")
print(df_sin_nulos_filas.head(10))

# Convertir cadenas a fechas
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    print("\n Columna 'Date' convertida a tipo datetime:")
    print(df['Date'].head(10))

# Eliminar duplicados
df_sin_duplicados = df.drop_duplicates()
print("\nDataFrame sin duplicados:")
print(df_sin_duplicados.head(10))
```

# 5. Transformación de Datos

```python
# Crea nuevas columnas basadas en operaciones con las columnas existentes
df['Total_Amount_Gs'] = df['Total Amount'] * 7500
print("\n DataFrame con la nueva columna 'Total_Amount_Gs':")
print(df[['Total Amount', 'Total_Amount_Gs']].head(10))

# Normalizar la columna 'Total_Amount_Gs'
max_value = df['Total_Amount_Gs'].max()
min_value = df['Total_Amount_Gs'].min()
df['Total_Amount_Gs_normalizado'] = df['Total_Amount_Gs'].apply(lambda x: (x - min_value) / (max_value - min_value))
print(df)

# Clasifica los datos en categorias relevantes
df['Status'] = df['Total_Amount_Gs_normalizado'].apply(lambda x: 'High' if x > 0.02 else 'Low')
print(df)
```

# 6. Análisis de Datos

```python
# Realiza agrupaciones de datos utilizando groupby para obtener insights específicos
ventas_por_genero = df.groupby('Gender').agg({
    'Age': 'mean',
    'Total Amount': 'sum'
})

print("\n Edad promedio y total gastado por género:")
print(ventas_por_genero)


# Aplica funciones de agregación como sum, mean, count, min, max, std, y var.
estadisticas_por_categoria = df.groupby('Product Category').agg({
    'Quantity': ['sum', 'mean', 'count', 'min', 'max', 'std', 'var'],
    'Price per Unit': ['sum', 'mean', 'min', 'max', 'std', 'var'],
    'Total Amount': ['sum', 'mean', 'min', 'max', 'std', 'var']
})

print("\n Estadísticas por categoría de producto:")
print(estadisticas_por_categoria)

# Definir una función personalizada para clasificar el nivel de gasto
def clasificar_cliente(fila):
    if fila['Total Amount'] > 2000:
        return 'Alto'
    elif fila['Total Amount'] > 1000:
        return 'Medio'
    else:
        return 'Bajo'

# Aplicar la función personalizada a cada fila usando 'apply'
df['Customer Spending Level'] = df.apply(clasificar_cliente, axis=1)

print("\nDataFrame con la nueva columna 'Customer Spending Level' (clasificación del nivel de gasto):")
print(df[['Total Amount', 'Customer Spending Level']].head(10))
```
