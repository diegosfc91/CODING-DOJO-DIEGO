# 2. Cargar los Datos

```python

import pandas as pd

# Cargar datos desde un archivo CSV
df = pd.read_csv('/content/CODING DOJO/retail_sales_dataset.csv')

# Ver las primeras y últimas filas del DataFrame
print("Primeras filas del DataFrame:")
df.head(10)
```

# 3. Transformación de Datos


```python
# Crea nuevas columnas
df['Total Revenue'] = df['Quantity'] * df['Price per Unit']
print("\nDataFrame con la nueva columna 'Total Revenue':")
print(df[['Quantity', 'Price per Unit', 'Total Revenue']].head(10))


# Clasifica los datos
df['Status'] = df['Total Revenue'].apply(lambda x: 'High' if x > 1000 else 'Low')
print(df)

# Aplica funciones de agregación como sum, mean, count, min, max, std, y var.
estadisticas_por_categoria = df.groupby('Product Category').agg({
    'Quantity': ['sum', 'mean', 'count', 'min', 'max', 'std', 'var'],
    'Price per Unit': ['sum', 'mean', 'min', 'max', 'std', 'var'],
    'Total Amount': ['sum', 'mean', 'min', 'max', 'std', 'var']
})

print("\n Estadísticas por categoría de producto:")
print(estadisticas_por_categoria)
```

# 4. Agrupación y Agregación

```python
# Agrupación por múltiples columnas
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month # Crear una nueva columna 'Month' a partir de la columna 'Date'

agrupacion_mes = df.groupby(['Month']).agg({
    'Total Amount': 'sum',
    'Quantity': 'sum'
}).reset_index() # Agrupar por'Month', y sumar 'Total Revenue' y 'Quantity'

print("\nAgrupación por 'Month' (suma de ingresos y cantidad):")
print(agrupacion_mes)

```
