```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas (corregido el uso de StringIO)
df = pd.read_csv(StringIO(data))

# Convertir la columna 'fecha' a formato datetime (corregido el nombre de la columna)
df['fecha'] = pd.to_datetime(df['fecha'])

# Gráfico de líneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o', linestyle='-', color='b')  # Añadido un marcador para mayor claridad
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)  # Añadir cuadrícula
plt.show()

# Crear un gráfico de dispersión
plt.figure(figsize=(10, 6))  # Asegurar un tamaño de figura adecuado
plt.scatter(df['fecha'], df['ventas'], color='g')
plt.title("Ventas Diarias (Dispersión)")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# Crear un histograma de las ventas
plt.figure(figsize=(10, 6))  # Asegurar un tamaño de figura adecuado
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
