```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
2023-01-07,230
2023-01-08,240
2023-01-09,260
2023-01-10,195
2023-01-11,280
"""

# Cargar los datos en un DataFrame de Pandas
df = pd.read_csv(StringIO(data))

# Convertir la columna 'fecha' a formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# An√°lisis Descriptivo
print("Estad√≠sticas Descriptivas:")
print(df['ventas'].describe())

# Visualizaci√≥n de la Distribuci√≥n de Ventas con Histogramas y Boxplots
plt.figure(figsize=(12, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Histograma de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(True)

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=df['ventas'], color='lightgreen')
plt.title("Boxplot de Ventas")
plt.ylabel("Ventas")
plt.grid(True)

plt.tight_layout()
plt.show()

# Gr√°fico de L√≠neas para mostrar la tendencia de ventas a lo largo del tiempo
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o', linestyle='-', color='b')
plt.title("Tendencia de Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Gr√°fico de Dispersi√≥n (utilizando el √≠ndice como eje x)
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['ventas'], color='g')
plt.title("Gr√°fico de Dispersi√≥n de Ventas")
plt.xlabel("√çndice")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# Combinaci√≥n de Histogramas y Boxplots
plt.figure(figsize=(12, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(df['ventas'], bins=5, alpha=0.5, color='skyblue', edgecolor='black', label='Histograma')
plt.title("Distribuci√≥n de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.legend()

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=df['ventas'], color='lightgreen', width=0.5)
plt.title("Distribuci√≥n de Ventas con Boxplot")
plt.ylabel("Ventas")

plt.tight_layout()
plt.show()

```
